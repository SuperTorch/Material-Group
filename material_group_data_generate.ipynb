# 生成模拟物资数据

```python
# 生成材料组的数据

import pandas as pd
import numpy as np

num = 100
columns_name = ['material_group_code', 'material_group_name']
df = pd.DataFrame(columns=columns_name)
df['material_group_code'] = range(1, num + 1)
df['material_group_name'] = ['材料组' + str(i) for i in range(1, num + 1)]
df
```

```python
# risk 24个字段, money 9 个字段
for i in range(24):
    temp = 'risk'+ str(i+1)
    columns_name.append(temp)
    # randint 正态分布（start，end，num）
    df[temp] = np.random.randint(1, 1000, num)
df   
```

```python
for i in range(9):
    temp = 'money'+ str(i+1)
    columns_name.append(temp)
    # randint 正态分布（start，end，num）
    df[temp] = np.random.randint(1, 1000, num)
df
```

```python
print(columns_name)
```

```python
df['risk_label'] = np.random.randint(0, 2, num)
df['money_label'] = np.random.randint(0, 2, num)
```

```python
risk_cols = ['risk' + str(i) for i in range(1,25)]
money_cols = ['money' + str(i) for i in range(1,10)]
```

```python
df['risk_score'] = 0
for column in risk_cols:
    df['risk_score'] += df[column]
df['money_score'] = 0
for column in money_cols:
    df['money_score'] += df[column]
df
```

```python
# 计算risk_score 和 money_score 的中位数
risk_median = df['risk_score'].median()
money_median = df['money_score'].median()
print(risk_median)
print(money_median)
```

```python
p = 0.95
for index, row in df.iterrows():
    if row['risk_score'] > risk_median:
        # 0 的概率为0.05，1 的概率为0.95
        df.loc[index, 'risk_label'] = np.random.choice([0,1], p = [1-p, p])
    else:
        df.loc[index, 'risk_label'] = np.random.choice([0,1], p = [p, 1-p])
    if row['money_score'] > risk_median:
        # 0 的概率为0.05，1 的概率为0.95
        df.loc[index, 'money_label'] = np.random.choice([0,1], p = [1-p, p])
    else:
        df.loc[index, 'money_label'] = np.random.choice([0,1], p = [p, 1-p])
```

```python
df.to_excel('./material_group2.xlsx',index=False)
```
